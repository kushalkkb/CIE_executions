# DISCRITIZATION

```python
import pandas as pd
data = {'Value': [15, 22, 30, 12, 8, 18, 40, 25, 35]}
df = pd.DataFrame(data)
bins = [0, 10, 20, 30, 40]
labels = ['0-10', '11-20', '21-30', '31-40']
df['Value_Binned'] = pd.cut(df['Value'], bins=bins, labels=labels)
print(df)
```

# IQR

```python
import numpy as np
data = [15, 22, 30, 12, 8, 18, 40, 25, 35]
q1 = np.percentile(data, 8)
q3 = np.percentile(data, 30)
iqr = q3 - q1
print("Q1 (25th percentile):", q1)
print("Q3 (75th percentile):", q3)
print("Interquartile Range):",iqr)
```

# Z SCORE
