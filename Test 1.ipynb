```python
import pandas as pd
dict={
    'Ename':['roshan','amar','ashwini','lohith','mohan','pramod'],
     'category':['regular','adhoc','regular','adhoc','contract','regular'],
     'department':['cs','cs','ec','mc','cs','ec'],
    'experience':['10','15','20','14','none','8'],
    'salary':['50000','15000','30000','15000','10000','40000']
}
df=pd.DataFrame(dict)
print(df)
```

```python
pivot_table = df.pivot_table(index='Ename', aggfunc='mean')
pivot_table
```

```python
from functools import reduce
def add(x,y):
    return(x+y)
number=(1,2,3,4,5)
result = reduce(add, number)
print(result)
```

```python
import pandas as pd
melted_df = pd.melt(df, id_vars=['Ename'], var_name='subject', value_name='salary')
melted_df
```

```python
number = (1,2,3,4,5)
def square(x):
    return x ** 2
square_number = map(square, number)
square_number_list = list(square_number)
print(square_number_list)
```

```python
number = (1,2,3,4,5)
def is_even(x):
    return x % 2 == 0
even_number = filter(is_even, number)
even_number_list = list(even_number)
print(even_number_list)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 2))
plt.bar(df['Ename'], df['salary'])
plt.xlabel('Ename')
plt.ylabel('Salary')
plt.title('bar graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
subject = 'salary'
plt.figure(figsize=(6, 6))
plt.pie(df[subject], labels=df['Ename'], autopct='%1.1f%%', startangle=140)
plt.title('pie chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.plot(df['Ename'], df['salary'], marker='o', linestyle='-', color='b')
plt.xlabel('Ename')
plt.ylabel('salary')
plt.title('line graph')
plt.show()
```
